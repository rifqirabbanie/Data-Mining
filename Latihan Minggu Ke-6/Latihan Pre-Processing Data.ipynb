```python
sydney['Humidity9am'] = sydney['Humidity9am'].fillna((sydney['Humidity9am'].mean()))
sydney['Humidity3pm'] = sydney['Humidity3pm'].fillna((sydney['Humidity3pm'].mean()))
sydney['Pressure9am'] = sydney['Pressure9am'].fillna((sydney['Pressure9am'].mean()))
sydney['Pressure3pm'] = sydney['Pressure3pm'].fillna((sydney['Pressure3pm'].mean()))
sydney['Cloud9am'] = sydney['Cloud9am'].fillna((sydney['Cloud9am'].mean()))
sydney['Cloud3pm'] = sydney['Cloud3pm'].fillna((sydney['Cloud3pm'].mean()))
sydney['Temp9am'] = sydney['Temp9am'].fillna((sydney['Temp9am'].mean()))
sydney['Temp3pm'] = sydney['Temp3pm'].fillna((sydney['Temp3pm'].mean()))
sydney['RISK_MM'] = sydney['RISK_MM'].fillna((sydney['RISK_MM'].mean()))

np.sum(sydney.isnull())
```

```python
from scipy import stats
sydney1=sydney.drop('Date',1)
sydney2=sydney1.drop('Location',1)
z = np.abs(stats.zscore(sydney2._get_numeric_data()))
print(z)
sydney = sydney[(z < 3).all(axis=1)]
print(sydney.shape)
```
