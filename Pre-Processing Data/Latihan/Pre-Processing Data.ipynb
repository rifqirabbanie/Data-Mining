# Praktikum Pre-Processing Data

```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv('G:\Kuliah\Data Mining\Praktikum\Pre-Processing Data\Latihan\movie_metadata.csv')
movie.head()
```

```python
movie.describe()
```

```python
np.sum(movie.isnull())
```

```python
movie['budget']=movie['budget'].fillna(3.975262e+07)
movie['duration']=movie['duration'].fillna(107.201074)
movie['gross']=movie['gross'].fillna(4.846841e+07)
movie['director_facebook_likes']=movie['director_facebook_likes'].fillna(686.621709)
movie['facenumber_in_poster']=movie['facenumber_in_poster'].fillna(1.371446)
movie['num_user_for_reviews']=movie['num_user_for_reviews'].fillna(272.770808)
movie['num_critic_for_reviews']=movie['num_critic_for_reviews'].fillna(140.194272)
movie['actor_1_facebook_likes']=movie['actor_1_facebook_likes'].fillna(6561.323932)
movie['actor_2_facebook_likes']=movie['actor_2_facebook_likes'].fillna(1652.080533)
movie['actor_3_facebook_likes']=movie['actor_3_facebook_likes'].fillna((movie['actor_3_facebook_likes'].mean()))

np.sum(movie.isnull())
```

```python
from scipy.stats import mode

movie['country']=movie['country'].fillna(movie['country']).dropna().mode()[0]
```

```python
movie.isnull().sum()
```

```python
#Mengatasi Outlier
from scipy import stats
movie1=movie.drop('country',1)
z = np.abs(stats.zscore(movie1._get_numeric_data()))
print(z)
movie = movie[(z < 3).all(axis=1)]
print(movie.shape)
```
