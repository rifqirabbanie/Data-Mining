```python
import pandas as pd
import numpy as np
```

```python
nfa = pd.read_csv(r'G:\Kuliah\Data Mining\Pre-processing Data\Tugas\NFA 2018.csv')
nfa.head()
```

```python
nfa.describe()
```

```python
np.sum(nfa.isnull())
```

```python
nfa['crop_land'] = nfa['crop_land'].fillna((nfa['crop_land'].mean()))
nfa['grazing_land'] = nfa['grazing_land'].fillna((nfa['grazing_land'].mean()))
nfa['forest_land'] = nfa['forest_land'].fillna((nfa['forest_land'].mean()))
nfa['fishing_ground'] = nfa['fishing_ground'].fillna((nfa['fishing_ground'].mean()))
nfa['built_up_land'] = nfa['built_up_land'].fillna((nfa['built_up_land'].mean()))
nfa['carbon'] = nfa['carbon'].fillna((nfa['carbon'].mean()))
nfa['Percapita GDP (2010 USD)'] = nfa['Percapita GDP (2010 USD)'].fillna((nfa['Percapita GDP (2010 USD)'].mean()))

np.sum(nfa.isnull())
```

```python
from scipy.stats import mode
from scipy import stats

z = np.abs(stats.zscore(nfa._get_numeric_data()))
print(z)
nfa = nfa[(z < 3).all(axis=1)]
print(nfa.shape)
```
